```python
import json
import matplotlib.pyplot as plt

# Load the JSON data from the file
with open('testData2.json', 'r') as file:
    data = json.load(file)

# Extract the timeseries data
timeseries = data['properties']['timeseries']

# Extract time and temperature data
time = [entry['time'] for entry in timeseries]
temperature = [entry['data']['instant']['details']['air_temperature'] for entry in timeseries]

# Plot the temperature data
plt.figure(figsize=(10, 5))
plt.plot(time, temperature, marker='o')
plt.title('Air Temperature Over Time')
plt.xlabel('Time')
plt.ylabel('Air Temperature (Celsius)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

# Save the plot as an image file
plt.savefig('temperature_plot.png')

# Display the plot
plt.show()
```
