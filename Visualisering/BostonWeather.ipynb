```python
import pandas as pd
import matplotlib.pyplot as plt

# Les inn datafilen
data = pd.read_csv("/Users/evinehfagerhaug/Library/Mobile Documents/com~apple~CloudDocs/Downloads/TDT4114/Miljo_data_analyse/data/BostonData2.csv")

# Fyll inn manglende verdier med gjennomsnitt
data.fillna(data.mean(numeric_only=True), inplace=True)

# Konverter 'time' til datetime-format hvis det ikke allerede er det
data['time'] = pd.to_datetime(data['time'])

# Sett tid som indeks
data.set_index('time', inplace=True)

# Plot temperaturutvikling
data['tavg'].plot(figsize=(12, 5), title="Gjennomsnittlig temperatur over tid", ylabel="Temperatur (°C)", xlabel="Dato")
plt.show()

# Plot nedbørsmengde
data['prcp'].plot(kind='bar', figsize=(12, 5), title="Daglig nedbør", ylabel="Nedbør (mm)", xlabel="Dato")
plt.show()

# Plot vindhastighet
data['wspd'].plot(figsize=(12, 5), title="Vindhastighet over tid", ylabel="Vindhastighet (km/t)", xlabel="Dato")
plt.show()

# Gjennomsnittstemperatur per måned
monthly_avg = data.resample('M').mean()
monthly_avg['tavg'].plot(figsize=(12, 5), title="Månedlig gjennomsnittstemperatur", ylabel="Temperatur (°C)", xlabel="Måned")
plt.show()

# Histogram for temperaturfordeling
data['tavg'].hist(bins=30, edgecolor='black', alpha=0.7)
plt.title("Temperaturfordeling")
plt.xlabel("Temperatur (°C)")
plt.ylabel("Antall dager")
plt.show()

print("Analyse fullført. Grafene er vist:)")
```
